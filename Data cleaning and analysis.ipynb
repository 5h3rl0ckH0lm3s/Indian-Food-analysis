```python
food_df.drop_duplicates(inplace = True) #removing duplicate columns
```

```python
food_df.duplicated().sum()
```

```python
food_df['prep_time'] = food_df['prep_time'].replace({-1 : food_df.loc[food_df['prep_time'] != -1, 'prep_time'].mean()})
food_df['cook_time'] = food_df['cook_time'].replace({-1 : food_df.loc[food_df['cook_time'] != -1, 'cook_time'].mean()})
```

```python
(food_df['prep_time'] == -1).sum()
```

```python
food_df['state'] = food_df['state'].replace({'-1' : 'Pan India'})
food_df['region'] = food_df['region'].replace({'-1' : 'India'})
```

```python
food_df[food_df['state'] == 'Pan India'].head()
```

```python
food_df['flavor_profile'] = food_df['flavor_profile'].replace({'-1' : 'mild'})
```

```python
(food_df['flavor_profile'] == '-1').sum()
```

```python
food_df['diet'] = food_df['diet'].replace({
    'veg': 'vegetarian',
    'non veg': 'non vegetarian'
})
```

```python
(food_df['diet'] == 'veg').sum()
```

```python
food_df.insert(food_df.columns.get_loc('ingredients') + 1, 'ingredient_count', food_df['ingredients'].str.split(',').str.len())
# creating a new column `ingredient_count` containing the number of ingredients next to the `ingredients` column.
```

```python
# Univariate Analysis started
plt.figure(figsize=(6, 4))
sns.histplot(food_df['prep_time'], bins=20, kde=False, edgecolor='black')
plt.title("Histogram of Prep Time")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(y=food_df['prep_time'])
plt.title("Box Plot of Prep Time")
plt.show()
```

```python
Q1_prep = food_df['prep_time'].quantile(0.25)
Q3_prep = food_df['prep_time'].quantile(0.75)
IQR_prep = Q3_prep - Q1_prep
outliers_prep = ((food_df['prep_time'] < (Q1_prep - 1.5 * IQR_prep)) | (food_df['prep_time'] > (Q3_prep + 1.5 * IQR_prep))).sum()
print(f"Outliers in Prep Time: {outliers_prep}")
```
